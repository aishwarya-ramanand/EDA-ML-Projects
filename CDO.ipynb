### **Running a Kruskal–Wallis Test**

-- We run a Kruskal–Wallis Test because there are more than 3+ groups in the CD_By_Suicide column, which is our target variable.

```python
from scipy.stats import kruskal
```

```python
cdo["CD_By_Suicide"] = cdo["CD_By_Suicide"].fillna(0)
```

```python
# Grouping the data by year

groups = [group["CD_By_Suicide"].values
          for year, group in cdo.groupby("Year")]
```

```python
stat, p = kruskal(*groups)
print("Statistic:", stat)
print("p-value:", p)
```

```python
cdo["Suicide_Occurred"] = cdo["CD_By_Suicide"].apply(lambda x: 1 if x > 0 else 0)
```

```python
cdo["Suicide_Occurred"]
```

```python
cdo["Suicide_Occurred"].value_counts()
```

```python
contingency = pd.crosstab(cdo["Area_Name"], cdo["Suicide_Occurred"])
contingency
```

```python
from scipy.stats import chi2_contingency

chi2, p, dof, expected = chi2_contingency(contingency)

print("Chi-square statistic:", chi2)
print("p-value:", p)
```

Chi-Square Test of Independence (State vs Suicide Occurrence)

A Chi-square test of independence was conducted to examine the relationship between State (Area_Name) and Suicide Occurrence (0 = No, 1 = Yes).
	•	Chi-square statistic: 93.20
	•	p-value: 3.84 × 10⁻⁸m

Since the p-value is significantly less than 0.05, we reject the null hypothesis and conclude that:

There is a statistically significant association between State and Suicide Occurrence.
This indicates that suicide rates in custody are not uniformly distributed across states; some states experience suicides at much higher frequencies than others.

### **Model Building**

```python
import pandas as pd
import numpy as np
```

```python
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import (
    accuracy_score, classification_report, confusion_matrix,
    roc_auc_score, roc_curve, precision_recall_fscore_support           
)
import matplotlib.pyplot as plt
import seaborn as sns
import joblib
```

#### **Quick checks & assertions**

```python
assert "Suicide_Occurred" in cdo.columns, "Create Suicide_Occurred target first."
print("Class distribution:\n", cdo["Suicide_Occurred"].value_counts())
```

#### **Feature Selection**

```python
cause_cols = [
    "CD_Accidents",
    "CD_By_Mob_AttackRiots",
    "CD_By_other_Criminals",
    "CD_IllnessNatural_Death",
    "CD_While_Escaping_from_Custody"
]

cat_cols = ["Area_Name", "Group_Name", "Sub_Group_Name"]
num_cols = ["Year"] + cause_cols

features = num_cols + cat_cols
print("Using features:", features)
```

```python
missing = [c for c in features if c not in cdo.columns]
if missing:
    raise ValueError("Missing columns in dataframe: {}".format(missing))

X = cdo[features].copy()
y = cdo["Suicide_Occurred"].copy()
```

```python
X[num_cols] = X[num_cols].fillna(0)

for c in cat_cols:
    X[c] = X[c].fillna("Unknown")
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20, random_state=42, stratify=y
)
```

```python
ohe = OneHotEncoder(handle_unknown="ignore", sparse_output=False)
preprocessor = ColumnTransformer(
    transformers=[
        ("num", StandardScaler(), num_cols),
        ("cat", ohe, cat_cols),
    ],
    remainder="drop",
    verbose_feature_names_out=False
)
```

```python
#Baseline: Logistic Regression pipeline
```

```python
logpipe = Pipeline(
    steps=[
        ("pre", preprocessor),
        ("clf", LogisticRegression(max_iter=1000, class_weight="balanced", random_state=42))
    ]
)
```

```python
logpipe.fit(X_train, y_train)
y_pred_log = logpipe.predict(X_test)
y_proba_log = logpipe.predict_proba(X_test)[:, 1]
```

```python
#  Evaluate Logistic Regression
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred_log))
print(classification_report(y_test, y_pred_log))
print("ROC-AUC:", roc_auc_score(y_test, y_proba_log))
```

Given a random suicide case and a random non-suicide case,
the model ranks the suicide case higher 78.1% of the time.

#### **Random Forest pipeline**

```python
rfpipe = Pipeline(
    steps=[
        ("pre", preprocessor),
        ("clf", RandomForestClassifier(n_estimators=200, class_weight="balanced", random_state=42))
    ]
)

rfpipe.fit(X_train, y_train)
y_pred_rf = rfpipe.predict(X_test)
y_proba_rf = rfpipe.predict_proba(X_test)[:, 1]
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred_rf))
print(classification_report(y_test, y_pred_rf))
print("ROC-AUC:", roc_auc_score(y_test, y_proba_rf))
```

```python
def plot_confusion(cm, labels=[0,1], title="Confusion matrix"):
    plt.figure(figsize=(4,3))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=labels, yticklabels=labels)
    plt.xlabel("Predicted"); plt.ylabel("Actual"); plt.title(title)
    plt.show()

cm_log = confusion_matrix(y_test, y_pred_log)
cm_rf = confusion_matrix(y_test, y_pred_rf)

plot_confusion(cm_log, title="Logistic Regression Confusion Matrix")
plot_confusion(cm_rf, title="Random Forest Confusion Matrix")
```

```python
fpr_log, tpr_log, _ = roc_curve(y_test, y_proba_log)
fpr_rf, tpr_rf, _ = roc_curve(y_test, y_proba_rf)

plt.plot(fpr_log, tpr_log, label=f"Logistic (AUC={roc_auc_score(y_test,y_proba_log):.3f})")
plt.plot(fpr_rf, tpr_rf, label=f"RandomForest (AUC={roc_auc_score(y_test,y_proba_rf):.3f})")
plt.plot([0,1],[0,1],"k--", alpha=0.5)
```

•	ROC curve visualizes tradeoff between True Positive Rate and False Positive Rate at many thresholds.
•	AUC near 0.5 = random; near 1.0 = excellent.
